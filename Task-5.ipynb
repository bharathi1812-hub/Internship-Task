```python
import pandas as pd
df = pd.read_csv('titanic_cleaned.csv', index_col=0)
```

```python
filtered_df=df[(df['Age']>=18)&(df['Fare']>30)]#filtering
filtered_df.head()
```

```python
df['Family_size']=df['sibsp']+df['Parch']+1 #feature1(creating new columns from existing columns-feature Engineering)
```

```python
df['FarePerPerson']=df['Fare']/df['Family_size']#feature2
```

```python
df.groupby('Pclass')['Fare'].mean()
```
