```python
import pandas as pd
df = pd.read_csv("titanic.csv")
```

```python
print(df.shape)
print(df.info())#to know about data like datatypes and null count
```

```python
print(df.head(4))
```

```python
df.isnull().sum()#checking null values and summing
```

```python
df.isna()#to check null values false=non-null,true=null
```

```python
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])#filling by mean
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)#to remove duplicates
df
```

```python
df.isnull().sum()#zero means all duplicates are removed
```

```python
df.duplicated().sum()
```

```python
df.dtypes#checking for wrong data type
```

```python
df.rename(columns={'2urvived': 'Survived'}, inplace=True)
```

```python
df = df[['Passengerid', 'Age', 'Fare', 'Sex', 'sibsp', 'Parch', 'Pclass', 'Embarked', 'Survived']]#keeping only important columns
df
```
